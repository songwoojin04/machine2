```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder

weather = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/3주차/weather.csv")
weather
```

```python
weather.isnull().sum()
```

```python
weather = weather.dropna()
weather.isnull().sum()
```

```python
weather.columns
```

```python
weather = weather.drop("RainTomorrow", axis=1)
```

```python
label_encoders = {}
for column in weather.columns:
    if weather[column].dtype == 'object':
        le = LabelEncoder()
        weather[column] = le.fit_transform(weather[column])
        label_encoders[column] = le
```

```python
weather.columns
```

```python
X = weather.drop('Rainfall',axis=1)
X.head()
```

```python
y = weather['Rainfall']
y.head()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, shuffle=True,random_state=12)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print(mean_squared_error(y_test, ly_preds))
```

해당 수의 의미는 테스트 데이터에서 예측값과 실제값의 차이를 제곱해서 평균 낸 값을 의미하며 값이 0에 가까울수록 모델이 잘 맞춘다는 의미입니다.

즉 예측값과 실제값 사이에 평균적으로 제곱 오차가 22.77
